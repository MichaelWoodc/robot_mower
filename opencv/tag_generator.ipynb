```python
import cv2
import numpy as np

# -----------------------------
# CONFIGURATION VARIABLES
# -----------------------------

DPI = 300
PAGE_WIDTH_IN = 8.5
PAGE_HEIGHT_IN = 11

# Convert inches → pixels
PAGE_W = int(PAGE_WIDTH_IN * DPI)
PAGE_H = int(PAGE_HEIGHT_IN * DPI)

# Margins (in inches)
MARGIN_IN = 0.5
MARGIN = int(MARGIN_IN * DPI)

# Tag size (in inches)
TAG_SIZE_IN = 2.0
TAG_SIZE = int(TAG_SIZE_IN * DPI)

# Spacing between tags (in inches)
SPACING_IN = 0.25
SPACING = int(SPACING_IN * DPI)

# AprilTag family
DICT = cv2.aruco.DICT_APRILTAG_36h11

# How many tags per row/column
ROWS = 3
COLS = 2

# Starting ID
START_ID = 0

# -----------------------------
# CREATE BLANK PAGE
# -----------------------------
page = np.ones((PAGE_H, PAGE_W), dtype=np.uint8) * 255  # white background

# Load dictionary
dictionary = cv2.aruco.getPredefinedDictionary(DICT)

# -----------------------------
# PLACE TAGS
# -----------------------------
tag_id = START_ID

for r in range(ROWS):
    for c in range(COLS):

        # Generate tag image
        tag = cv2.aruco.generateImageMarker(dictionary, tag_id, TAG_SIZE)

        # Compute placement coordinates
        x = MARGIN + c * (TAG_SIZE + SPACING)
        y = MARGIN + r * (TAG_SIZE + SPACING)

        # Paste tag into page
        page[y:y+TAG_SIZE, x:x+TAG_SIZE] = tag

        # Label under tag
        cv2.putText(page, f"ID {tag_id}", (x, y + TAG_SIZE + 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, 0, 3)

        tag_id += 1

# -----------------------------
# SHOW RESULT
# -----------------------------
cv2.imshow("AprilTag Sheet", page)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# -----------------------------
# CONFIGURATION
# -----------------------------

# AprilTag family
DICT = cv2.aruco.DICT_APRILTAG_36h11

# Tag ID to generate
TAG_ID = 0

# Output size (pixels)
# 1000–2000 px is ideal for printing
TAG_SIZE = 1200

# Output filename
OUTPUT_FILE = f"apriltag_{TAG_ID}.png"

# -----------------------------
# GENERATE TAG
# -----------------------------
dictionary = cv2.aruco.getPredefinedDictionary(DICT)

# Create the tag image
tag_img = cv2.aruco.generateImageMarker(dictionary, TAG_ID, TAG_SIZE)

# Save to current directory
cv2.imwrite(OUTPUT_FILE, tag_img)

print(f"Saved {OUTPUT_FILE} ({TAG_SIZE}x{TAG_SIZE}px)")
```
