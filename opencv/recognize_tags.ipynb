```python
import cv2
import numpy as np
import math

# -----------------------------
# KNOWN REAL VALUES
# -----------------------------
REAL_TAG_SIZE = 0.20      # meters
KNOWN_DISTANCE = 0.9144   # meters (3 ft)

# -----------------------------
# CAMERA
# -----------------------------
cap = cv2.VideoCapture(0)

dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_APRILTAG_36h11)
detector = cv2.aruco.ArucoDetector(dictionary)

calibrated_fx = None
camera_matrix = None
dist_coeffs = np.zeros((5, 1))

print("Press 'c' to calibrate at 3 ft, ESC to quit")

# 3D model points (updated after calibration)
obj_points = None

while True:
    ret, frame = cap.read()
    if not ret:
        break

    h, w = frame.shape[:2]

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    corners, ids, _ = detector.detectMarkers(gray)

    if ids is not None:
        pts = corners[0][0]

        # Draw tag
        cv2.polylines(frame, [pts.astype(int)], True, (0,255,0), 2)

        # -----------------------------
        # CALIBRATION MODE
        # -----------------------------
        pixel_width = np.linalg.norm(pts[0] - pts[1])

        cv2.putText(frame, f"Pixel Width: {pixel_width:.1f}",
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX,
                    0.7, (0,255,0), 2)

        fx_live = (pixel_width * KNOWN_DISTANCE) / REAL_TAG_SIZE
        cv2.putText(frame, f"Live fx: {fx_live:.1f}",
                    (20, 70), cv2.FONT_HERSHEY_SIMPLEX,
                    0.7, (0,255,0), 2)

        # -----------------------------
        # POSE MODE (after calibration)
        # -----------------------------
        if camera_matrix is not None:

            success, rvec, tvec = cv2.solvePnP(
                obj_points, pts, camera_matrix, dist_coeffs
            )

            if success:
                # Draw axis
                cv2.drawFrameAxes(frame, camera_matrix, dist_coeffs, rvec, tvec, REAL_TAG_SIZE/2)

                # Convert rotation matrix
                R, _ = cv2.Rodrigues(rvec)

                sy = math.sqrt(R[0,0]**2 + R[1,0]**2)
                singular = sy < 1e-6

                if not singular:
                    yaw = math.atan2(R[2,1], R[2,2])
                    pitch = math.atan2(-R[2,0], sy)
                    roll = math.atan2(R[1,0], R[0,0])
                else:
                    yaw = math.atan2(-R[1,2], R[1,1])
                    pitch = math.atan2(-R[2,0], sy)
                    roll = 0

                yaw = math.degrees(yaw)
                pitch = math.degrees(pitch)
                roll = math.degrees(roll)

                x, y, z = tvec.flatten()

                # Overlay pose info
                lines = [
                    f"X: {x:.2f}m Y: {y:.2f}m Z: {z:.2f}m",
                    f"Yaw: {yaw:.1f}",
                    f"Pitch: {pitch:.1f}",
                    f"Roll: {roll:.1f}",
                ]

                base_x, base_y = int(pts[0][0]), int(pts[0][1]) - 10

                for j, line in enumerate(lines):
                    cv2.putText(frame, line,
                                (base_x, base_y - j*20),
                                cv2.FONT_HERSHEY_SIMPLEX,
                                0.6, (0,255,0), 2)

    # Show resolution
    cv2.putText(frame, f"Resolution: {w}x{h}",
                (20, 110), cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (255,255,0), 2)

    # Show calibration result
    if calibrated_fx is not None:
        cv2.putText(frame, f"CALIBRATED fx = {calibrated_fx:.1f}",
                    (20, 150), cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (0,0,255), 2)

    cv2.imshow("Calibration + Pose", frame)

    key = cv2.waitKey(1) & 0xFF

    # -----------------------------
    # PRESS 'c' TO CALIBRATE
    # -----------------------------
    if key == ord('c') and ids is not None:
        calibrated_fx = (pixel_width * KNOWN_DISTANCE) / REAL_TAG_SIZE

        cx = w / 2
        cy = h / 2

        camera_matrix = np.array([
            [calibrated_fx, 0, cx],
            [0, calibrated_fx, cy],
            [0, 0, 1]
        ], dtype=np.float32)

        # Build 3D tag model
        half = REAL_TAG_SIZE / 2
        obj_points = np.array([
            [-half,  half, 0],
            [ half,  half, 0],
            [ half, -half, 0],
            [-half, -half, 0]
        ], dtype=np.float32)

        print("\n=== CALIBRATION RESULT ===")
        print(f"Resolution: {w}x{h}")
        print(f"fx = fy = {calibrated_fx:.2f}")
        print(f"cx = {cx:.2f}")
        print(f"cy = {cy:.2f}")
        print("==========================\n")

    elif key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
